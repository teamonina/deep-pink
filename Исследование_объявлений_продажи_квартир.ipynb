# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Описание данных:

- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах (м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

Сразу можно увидеть, что всего 22 столбца, узнаем более подробную информацию.

```python
df.info()
```

- Имеются пропуски, необходимо заполнить

## Предобработка данных

Для начала посмотрим, какой процент пропусков в каждом столбце

```python
df.isna().sum()/df.shape[0]*100
```

Предположим, что от некоторых пропусков можно избавиться, проанализируем столбцы, обратимся к числовому описанию данных.

```python
df.describe()
```

В столбце 'is_apartment' большое количество пропусков, значения в нём - булево. Пропуск-ложь, говорит о том, что недвижимость не является апартаментами, поэтому пропуски лучше заменить на 0. Аналогично,  в столбце 'balcony' пропуск может показывать отсутсвие балкона, в столбцах 'parks_around3000' и 'ponds_around3000' пропуск означает отсутствие парков и водоёмов около этой недвижимости, их тоже заменим на 0. <br> 

```python
df['is_apartment'] = df['is_apartment'].fillna(0)
df['balcony'] =  df['balcony'].fillna(0)
df['parks_around3000'] = df['parks_around3000'].fillna(0)
df['ponds_around3000'] =  df['ponds_around3000'].fillna(0)
```

Обратим внимание на столбцы 'parks_nearest' и 'ponds_nearest', в них пропусков больше чем в 'parks_around3000' и 'ponds_around3000', значит парка или водоёма, расстояние до которых менее 3000 - нет. <br>
Аналогично поступим со столбцами 'cityCenters_nearest' и 'airports_nearest': пропуск расценивается как отсутсвие аэропорта и центра города рядом.

```python
df['parks_nearest'] = df['parks_nearest'].fillna(0)
df['ponds_nearest'] =  df['ponds_nearest'].fillna(0)
df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna(0)
df['airports_nearest'] =  df['airports_nearest'].fillna(0)
```

В столбце 'ceiling_height' большое количество пропусков, притом минимальное значение - 1м, а максимальное - 100м. Очевидно, что здесь есть ошибка. Заменим пропуски здесь медианным значением, так как оно более устойчиво к вбросам.

```python
df['ceiling_height'] =  df['ceiling_height'].fillna(df['ceiling_height'].median())
```

Менее 1% у 'floors_total' и 'locality_name'. Сложно предположить город или количество этажей в доме, стандартное отклонение в первом случае - практически 7 этажей, а для категориального значения замену не создать. Придётся оставить пропуски. <br>
Для заполнения столбцов 'living_area' и 'kitchen_area' возьмём среднее соотношение жилой площади и площади кухни к общей площади недвижимости. <br>

```python
mean_living = (df['living_area']/df['total_area']).median()
mean_kitchen = (df['kitchen_area']/df['total_area']).median()
df['kitchen_area'] = df['kitchen_area'].fillna(mean_kitchen)
df['living_area'] = df['living_area'].fillna(mean_kitchen)
```

Проверим на критичные случаи. Не превышает ли кухня+жилое пространство общее пространство


```python
df.query('living_area + kitchen_area>total_area')
```

Выяснилось, что таких критичных случаев 52. В них заменим площадь кухни на разницу между общей площадью и жилой.

```python
def liv_kit(liv, kit, tot):
    if (liv + kit)>tot:
        return tot - liv
    return kit
df['kitchen_area'] = df.apply(lambda x: liv_kit(x.living_area,x.kitchen_area,x.total_area), axis=1)
```

Остался последний столбец, требующий корректировки. Выведем дату публикаций тех объявлений, которые имеют пропуск.

```python
df.loc[df['days_exposition'].isna()==True,['first_day_exposition','days_exposition']]
```

Для архива относительно свежие даты. Думаю, на момент составления датасета объявления еще были актуальны. Оставим эти пропуски как есть. А какой это тип данных? Возможно, его надо поменять, а заодно и проверить остальные столбцы.

```python
df.dtypes
```

Чтобы в дальнейшем оперировать датами, переведём 'first_day_exposition' из типа object в тип datetime.

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
```

Приведем колонки с типом float64 'floors_total','parks_around3000','ponds_around3000' и 'days_exposition' в тип int32 для экономии памяти


```python
list_type = ['floors_total','parks_around3000','ponds_around3000','days_exposition']
for i in df.columns:
    if i in list_type:
        df[i] = df[i].astype('int32', errors='ignore')
df.info()
```

### Вывод

Все пропуски, которые имели логичную замену были исправлены. <br>
- 'ceiling_height' - взята медианная высота потолков 
- 'floors_total', 'locality_name' - оставлены без изменений
- 'living_area', 'kitchen_area' - взято медианное соотношение к общей площади <br>
Остальные пропуски заменены на 0. <br>
В столбцу 'first_day_exposition' изменён тип данных на datetime. В остальных столбцах замена не требуется, с ними можно выполнять все операции.

## Расчёты и добавление результатов в таблицу

Добавим дополнительный столбец - цена квадратного метра 'price_per_meter'. Разделим цену на момент снятия с публикации на общую площадь.

```python
df['price_per_meter'] = df['last_price']/df['total_area']
df.head()
```

Добавим еще столбцы: 
 - день недели 'wday'
 - месяц 'month'
 - год 'year'

```python
df['wday'] = df['first_day_exposition'].dt.weekday
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
```

Добавим свойство этажей 'floor_category': 'Первый', 'Последний', 'Другой'. Для этого потребуется функция floor_cat_func

```python
def floor_cat_func(floor, total_floor):
    if floor==1:
        return 'Первый'
    if floor==total_floor:
        return 'Последний'
    return 'Другой'
df['floor_category'] = df.apply(lambda x: floor_cat_func(x.floor,x.floors_total), axis=1)
```

Добавим соотношение жилой и общей площади 'ratio_living', а также кухни и общей площади 'ratio_kitchen'

```python
df['ratio_living'] = df['living_area']/df['total_area']
df['ratio_kitchen'] = df['kitchen_area']/df['total_area']
```

Проверим наличие новых столбцов в таблице, сделаем срез первых 5 строк и новых столбцов.

```python
df.loc[:5,'price_per_meter':'ratio_kitchen']
```

### Вывод

Теперь в таблице есть новые столбцы 'price_per_meter', 'wday', 'month', 'year', 'floor_category',
       'ratio_living', 'ratio_kitche', 'ratio_kitchen'

### Изучим предложения квартир Санкт-Петербурга

Изучим Санкт-Петербург, задача состоит в том, чтобы выяснить, какая область входит в центр. Выделим квартиры в Санкт-Петербурге и создадим столбец, где расстояние до центра будет округлено до целого значения

```python
loc_sp = df.loc[df['locality_name']=='Санкт-Петербург',['locality_name','cityCenters_nearest','last_price']].reset_index()
loc_sp['cityCenters_nearest'] = loc_sp['cityCenters_nearest']/1000
loc_sp.head()
```

Округлим значения в километрах с небольшим запасом в большую сторону.

```python
import numpy as np
loc_sp['cityCenters_nearest'] = loc_sp['cityCenters_nearest'].apply(np.ceil)
loc_sp.head()
```

Добавим столбец, в котором будут посчитано среднее значение за км. Затем посчитаем среднее по столбцу.

```python
loc_sp['price_per_km']=loc_sp['last_price']/loc_sp['cityCenters_nearest']
display(loc_sp.head())
loc_sp['price_per_km'].median()
```

```python
loc_sp
```

Построим график, показывающий изменения цены от удаленности от центра.

```python
loc_sp.plot(x='cityCenters_nearest',y='price_per_km', style='o')
```

На графике видно, как с увеличением расстояния от центра падает цена. Попробуем построить график другого плана, где будут учитываться средние показатели цены.

```python
loc_sp.groupby('cityCenters_nearest')['last_price'].mean().plot()
```

Резкий спад виден после отметки в 5 км, значит центром называется область радиусом 5 км.

Сильной взаимосвязи между расстоянием до центра и ценой в центральном районе нет.
